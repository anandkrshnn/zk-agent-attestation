This notebook demonstrates the lifecycle of an **AI Agent Identity Proof** using the PTV protocol. We will walk through collecting hardware-anchored claims, transforming them into a Zero-Knowledge Proof (ZKP), and verifying the proof on a remote orchestrator.

## 1. Environment Setup
Installing core PTV dependencies.

```python
# !pip install pyyaml requests pycryptodome
import sys
import os
import time
import json
from datetime import datetime

# Simulation of the PTV local environment
sys.path.append('..')
from src.utils import calculate_sha256, log
from src.prove import collect_claims, generate_proof
from src.verify import verify_proof

print(f"PTV Environment Initialized: {datetime.now()}")
```

## 2. Phase 1: Prove (Claims Collection)
In this phase, the agent collects state measurements from the local hardware (TPM 2.0 / Secure Enclave).

```python
model_id = "Llama-3-8B-Healthcare"
policy_path = "config/icu_safety_v2.yaml"

print("[*] Initiating hardware-anchored measurement...")
claims = collect_claims(model_id, policy_path)

print("\n--- PROVE RESULTS ---")
print(json.dumps(claims, indent=2))
```

## 3. Phase 2: Transform (ZKP Generation)
The raw claims are transformed into a privacy-preserving Groth16 zk-SNARK proof.

```python
print("[*] Starting Zero-Knowledge transformation...")
proof = generate_proof(claims)

print("\n--- TRANSFORM RESULTS ---")
print(f"Protocol: {proof['protocol']}")
print(f"Proof: {proof['proof']}")
print(f"Public Signals: {proof['public_signals']}")
```

## 4. Phase 3: Verify (Remote Validation)
The orchestrator verifies the proof. This check is fast (O(1)) and occurs before any institutional data access.

```python
expected_signals = [claims['model_hash'], claims['policy_fingerprint']]

print("[*] Verifying proof at remote gateway...")
is_valid = verify_proof(proof, expected_signals)

if is_valid:
    print("\n✅ SUCCESS: Agent identity authenticated. Trusted handoff initiated.")
else:
    print("\n❌ FAILURE: Proof invalid.")
```
